# Web Component Performance Testing: ImageSlider

The previous chapter analyzed a rather simple web component with only one attribute.  

The ImageSlider web component, which will be tested in the following chapter is a more difficult component.   
The component accepts images, which will be reflected as the different slides of the slider.
Furthermore, the initial visible slide can be set via the `selectedindex` attribute. 
Click events will be fired when a click on a dot is performed or the arrow keys are used. 
The component fires the event `updateSelectedIndex` for noticing the parent component about the change. 
The parent component may use this information for reflecting to the `selectedindex` attribute, emitting an update in the ImageSlider component. The newly set index will lead to a slide change in the web component.



example of using the ImageSlider component inside an HTML page

---

```html
    <image-slider selectedindex="0">
        <img src="/static/img/pic-1.jpg">
        <img src="/static/img/pic-2.jpeg">
        <img src="/static/img/pic-3.jpeg">
        <img src="/static/img/pic-4.jpeg">
    </image-slider>
    <script>
       const ImageSlider = document.querySelector('image-slider');
        ImageSlider.addEventListener('updateSelectedIndex', (ev)=>{
            ImageSlider.setAttribute('selectedindex', ev.detail)
        })
    </script>
```



## Performance Measurement
The previous chapter already explained the `check_performance_new_boot` function and why it is important not only to just reload the browser window, but reboot the browser entirely for getting confidential results.
Here this function will be reused.

Via JavaScript the component slides from slide at position zero over the second until the third slide. The passed time is persisted in the LocalStorage.

````python
#function from the previos chapter

chromedriver = "/usr/local/bin/chromedriver"


def check_performance_new_boot(port=8000, sleep=2, page_calls=100,
							   foldername='00_web-component', 
							   path='performance_test_result/00_image-slider'):
	"""opens chrome browser multiple times 
        and saves the token contents from 'bodyRendering' into a bodyRender.json file
	    
	Args:
		port: Number defining on which port of localhost the application is running on
		sleep: time in seconds before page reload
		page_calls: amount of times the URL is requested
		foldername: name of the folder, where to store the resulting JSON file
		path: relative path to foldername
	"""
    
	# first we define the url of the application 
	# and initialize our resulting performance array
	url = 'http://127.0.0.1:' + str(port)
	result = []
	
	# in a loop we open and close the chrome browser and get the specified url
	# for making sure the page was rendered completely and the LocalStorage was filled 
		# we wait some seconds before we add the single value in the LocalStorage array 
		# to our resulting performance array
	for i in range(page_calls):
		driver = webdriver.Chrome(chromedriver)
		driver.implicitly_wait(100)
		driver.get(url)
		time.sleep(sleep)
		result.append(pd.read_json((
            driver.execute_script("return localStorage.getItem('bodyRendering')")
        )).values[0][0])

		driver.quit()
	
	local_path = path
	
	#get current working directory
	path = os.getcwd() + '/' + path + '/' + foldername
	
    # write the resulting JSON file
	f = open(path + "/BodyRendering.json","w")
	f.write(str(result))
	f.close()
	print('ü•≥  üéâ üçæ Yay, successfully created "'+ local_path + '/' + foldername + '/BodyRendering.json"' )

	
    # return the mean value of our resulting performance array
	return np.mean(result)
	
		

````

The next section iterates over the component files by calling the `check_performance_new_boot` function for each component test. The function then stores the measurements in a `BodyRendering.json` file.

````python
# iterate over web components
foldernames = ['00_web-component',  
               '01-angular', 
               '02-polymer', 
               '03-vue']

[check_performance_new_boot(page_calls=100,sleep=5, foldername=foldernames[i], port=8000+i) for i in range(len(foldernames))]
````

This section reads in the measurements stored in the JSON files and puts the calculations into a Pandas DataFrame.

````python
body_rendering_path = 'performance_test_result/00_image-slider/'

# read the json files in with pandas
web_component_dataframe = pd.read_json(body_rendering_path + '00_web-component/BodyRendering.json')
angular_dataframe = pd.read_json(body_rendering_path + '01-angular/BodyRendering.json')
polymer_dataframe = pd.read_json(body_rendering_path + '02-polymer/BodyRendering.json')
vue_dataframe = pd.read_json(body_rendering_path + '03-vue/BodyRendering.json')

# create a dataframe for the plain components and set the column names
frames = [web_component_dataframe, angular_dataframe, polymer_dataframe, vue_dataframe]
performance_columns = ['web_component', 'angular', 'polymer', 'vue']
performance_dataframe = pd.concat(frames, axis=1, sort=False)
performance_dataframe.columns = performance_columns



performance_dataframe.describe()
````

The description shows a rather high standard deviation, which is why a graph may help to identify spikes.

````python
# figure for the web components created with the frameworks
f = plt.figure()
performance_dataframe.plot(kind='line',figsize=(14,7), title="Framework Analysis",
                                       grid=True,stacked = False)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel("ms")
plt.xlabel("app call")
plt.show()
````

As expected the graph shows some spikes. For flattening the lines the previously defined function `regulate_df` is reused.

````python
# define function for flattening the performance measurement graphs
def regulate_df(df):
	mean = df[0].mean()
	std = df[0].std()
    # replace outliers that are outside the [mean-std, mean+std] area 
    # with mean values
	df[0] = np.where(df[0] < mean - std, mean,df[0])
	df[0] = np.where(df[0] > mean + std, mean,df[0])
	return df
````

````python
# iterate over web component DateFrames
for df in frames:
	regulate_df(df)
````

````python
# check new description of the regulated DataFrame
frames = [web_component_dataframe, angular_dataframe, polymer_dataframe, vue_dataframe]
performance_columns = ['web_component', 'angular', 'polymer', 'vue']
performance_dataframe = pd.concat(frames, axis=1, sort=False)
performance_dataframe.columns = performance_columns
performance_dataframe.describe()
````

````python
# figure for the web components created with the frameworks
f = plt.figure()
performance_dataframe.plot(kind='line',figsize=(14,7), title="Framework Analysis",
                                       grid=True,stacked = False)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.ylabel("ms")
plt.xlabel("app call")
plt.show()
````

## Observations
The Graph shows that the web component created with the help of the Polymer library performs best and the one created with Angular worst. 

- Angular performs about 5% worse then polymer and about 3% worse then the vanilla web component.
- The vanilla Web Component performs about 2.5% worse then the Polymer web component.
- Vue performs about .7% worse then the Polymer web component.

However, the graph displays huge differences, taking a look at the comparison percentages, the difference in performance is only between .7% and 5%, which is not that height, but could be important when using many components on a website.

# Conclusion and Further Thoughts

The creation of plain Web Components is very well documented by [the Mozilla MDM web docs](https://developer.mozilla.org/en-US/docs/Web/Web_Components) or the[ Google developer documentation](https://developers.google.com/web/fundamentals/web-components/).

The creation of the components with slots, especially applying styling via the `::slotted` attribute was neither possible in Vue nor Angular and could only be performed by binding the styles directly to the HTML nodes.

However, Polymer works by far better with styling, the Polymer 3 documentation is still little. To solve the issue how a parameter could be passed if the user clicks on a dot, came with a great research process.

All in all, which one to choose is about personal preference of course. However, as current recommendation, Angular for components build for outside use of the framework, is not the most fitting one, since the slow performance and the missing support for slots. A lot of Angular web components could reduce the website's speed drastically compared to the contestants.

To create a web component with Vue is very comfortable and not different to when just creating usual Vue.js components. Only the `::slotted` attribute came with little difficulties, which via style binding were easily resolved. Furthermore, this component consists of the fewest lines of code a developer has to write.


Nevertheless, one major issue that was not discussed yet are common styles.
Some styles like the font styles or theme specific styles that make the website look consistent currently can only be performed with great effort. 

One could for example include a general stylesheet into all the components. However, this would lead to a lot of overhead and make sharing the Web Components more difficult.

Or an attribute could be defined for the components. E.g. a font attribute etc.

The perfect solution here would be to declare inside the component, which stylings should be applied from the hosting Web Page. 

Before this issue is not solved, simple shareable Web Components, that can be easily plugged into a HTML Web Page, will probably not appear to much. 


Components will likely exist in one application which is bound to one framework only. Hopefully in the future this issue will be solved and collaboration becomes easier.
